```python
import pandas as pd
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/market.csv')
print(df.to_string())
```

```python
# do association rule mining
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_bought = apriori(df, min_support = 0.1, use_colnames=True)
print("list of frequent bought items")
frequent_bought
```

```python
rules = association_rules(frequent_bought, metric = "confidence", min_threshold = 0.65)
rules
```
